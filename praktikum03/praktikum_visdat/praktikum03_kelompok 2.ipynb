Import library dan dataset

```python
import matplotlib.pyplot as plt

# Data sample
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10, 20, 15, 25, 30, 45, 40, 50, 60, 55, 65, 70]
product_B_sales = [5, 10, 8, 15, 18, 20, 22, 30, 25, 35, 40, 45]

# line plot
plt.plot(month, product_A_sales, label='Product A')
plt.title('Penjualan produk A per bulan')
plt.xlabel('Bulan')
plt.legend()
plt.grid(True)
plt.ylabel('Jumlah penjualan')
plt.show()
```

multiple line plot & customizations

```python
# line plot
plt.plot(month, product_A_sales, label='Product A',linestyle="--", marker='o', color='blue')
plt.plot(month, product_B_sales, label='Product B',linestyle="--", marker='x', color='red')

plt.title('Penjualan produk per bulan')
plt.xlabel('Bulan')
plt.legend()
plt.grid(True)
plt.ylabel('Jumlah penjualan')
plt.show()
```

**jenis garis untuk menunjukan tren**

```python
# data sample
product_C_sales = [18, 22, 25, 28, 32, 38, 42, 45, 48, 52, 56, 60]
product_D_sales = [7, 9, 11, 13, 16, 18, 20, 23, 25, 28, 30, 35]

# plot product A
plt.plot(month, product_A_sales, label='Product A',linestyle="--", color='blue')

# plot product B
plt.plot(month, product_B_sales, label='Product B',linestyle="-", color='red')

# plot product C
plt.plot(month, product_C_sales, label='Product C',linestyle=":", color='green')

# plot product D
plt.plot(month, product_D_sales, label='Product D',linestyle="-.", color='purple')

plt.title('Penjualan produk per bulan')
plt.xlabel('Bulan')
plt.legend()
plt.grid(True)
plt.ylabel('Jumlah penjualan')
plt.show()
```

**sublot pada lice chart**

```python
# subplot, 2 baris 1 kolom, panjang 10 inch, tinggi 8 inchi

fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk produk C
axs[0].plot(month, product_C_sales, label='Product C',linestyle=":", color='green', marker='d')
axs[0].set_title('Penjualan produk C per bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah penjualan')
axs[0].legend()
axs[0].grid(True)

# plot pertama untuk produk D
axs[1].plot(month, product_D_sales, label='Product D',linestyle=":", color='purple', marker='s')
axs[1].set_title('Penjualan produk D per bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
