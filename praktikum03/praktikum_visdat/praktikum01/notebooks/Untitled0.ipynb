```python
import matplotlib.pyplot as plt
import pandas as pd

# Data dummy pengeluaran APBN
data = {'Kategori': ['Pendidikan', 'Kesehatan', 'Infrastruktur', 'Pertahanan', 'Lainnya'],
        'Pengeluaran (dalam triliun Rupiah)': [20, 15, 25, 10, 30]}
df = pd.DataFrame(data)

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Pengeluaran (dalam triliun Rupiah)'], labels=df['Kategori'], autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Pengeluaran APBN (Data Dummy)')
plt.axis('equal')  # Memastikan pie chart berbentuk lingkaran
plt.show()
```
