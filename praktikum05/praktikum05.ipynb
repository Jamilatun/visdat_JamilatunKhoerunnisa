# Praktikum 5 Visualisasi data - Scatter plot

1. import ilbrary

```python
import matplotlib.pyplot as plt
```

2. buat dataset

```python
# membuat data
suhu =  [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]
```

3. Basic scatter plot

```python
plt.scatter(suhu, penjualan)
plt.title('Hubungan penjualan Eskrim dan suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Eskrim')
plt.show()
```

4. kustomisasi scatter plot

```python
plt.scatter(suhu, penjualan, color='orange', s=100, edgecolors='black', alpha=0.7)

plt.title('Hubungan penjualan Eskrim dan suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Eskrim')
plt.grid(True)
plt.show()
```

5. multiple scatter plot

```python
# dataset suhu
suhu =  [20, 22, 24, 26, 28, 30, 32, 34, 36]

# dataset penjualan
penjualan_weekday = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

plt.scatter(suhu, penjualan_weekday, color='green', label='Hari kerja', s=80)
plt.scatter(suhu, penjualan_weekend, color='purple', label='Akhir pekan', s=80)

plt.scatter(suhu, penjualan)
plt.title('Hubungan penjualan Eskrim dan suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Eskrim')
plt.legend()
plt.show()
```

6. melihat hubungan data dengan beberapa variabel

```python
suhu =  [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# data tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [200, 220, 240, 260, 280, 300, 320, 340, 360]

plt.scatter(suhu, penjualan, c=kelembapan, s=ukuran_titik, cmap='coolwarm', alpha=0.6)
plt.title('Hubungan penjualan kelembapan dan suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Eskrim')
plt.colorbar(label='Kelembapan')
plt.show()
```

7. 3D scatter plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# dataset suhu
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# dataset penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

# 3D Scatter Plot
fig = plt.figure(figsize=(15, 13))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(suhu, penjualan_weekdays, penjualan_weekend,
           color='blue', s=80, label='Data Penjualan')

ax.set_title('Hubungan Penjualan Es Krim dan Suhu (3D ver)')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Hari Kerja')
ax.set_zlabel('Penjualan Akhir Pekan')
ax.legend()
plt.show()
```
